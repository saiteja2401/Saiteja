## Classification Naive Bayes

```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#importing data set
train_data = pd.read_csv('har_train.csv')
test_data = pd.read_csv('har_validate.csv')
```

```python
#train and test data shape
print('train shape - ', str(train_data.shape))
print('test shape - ', str(test_data.shape))
```

```python
#Insights of data
train_data.info()
```

```python
#Data summary
train_data.describe()
```

```python
#visualizing the classes in target variable
sns.countplot(train_data.classe)
```

```python
#Correlation
copy_data = train_data.copy()
train_data.corr()
```

```python
#Splitting the data into train and test
X_train = train_data.drop('classe', axis = 1)
y_train = train_data['classe']

X_test = test_data.drop('classe', axis = 1)
y_test = test_data['classe']
```

```python
#Scaling the features
mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
X_test = mms.transform(X_test)
```

```python
#Fitting the model
model = GaussianNB()

model.fit(X_train, y_train)
```

```python
#Prediction
y_pred = model.predict(X_test)
```

```python
#Accuracy of train and test
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
#Confusion Matrix
plt.figure(figsize=(12, 8))
ax =sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cbar = False, square = True, 
            xticklabels=y_test.unique(), yticklabels=y_test.unique(), fmt = 'd')
ax.set_xlabel('Predict',fontsize = 15)
ax.set_ylabel('Actual',fontsize = 15)
```

```python
#Classification report
print(classification_report(y_test, y_pred))
```

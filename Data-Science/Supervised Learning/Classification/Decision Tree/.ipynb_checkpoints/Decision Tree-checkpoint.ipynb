## Classification Using Decision Tree

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading the data from csv
data = pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
#Data info
data.info()
```

```python
#Summary of data
data.describe()
```

```python
#Visualizing the data
sns.countplot(data.Purchased)
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data.Age, hue = data.Purchased)
```

```python
#Relation between the features
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), fmt= '.2f', square = True, cmap = 'Blues', annot = True)
```

As we can see that userID is weekly correlated with dependent.

```python
#Dropping UserId column
data.drop('User ID', inplace=True, axis = 1)
```

```python
#Checking the null values
data.isnull().sum()
```

```python
#encoding the categorical values
data = pd.get_dummies(data, drop_first=True)
```

```python
#Splitting the data into dependent and independent
X = data.drop('Purchased', axis = 1)
y = data.Purchased
```

```python
#Train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101) 
```

```python
#Model building
model = DecisionTreeClassifier()# splitting on gini criteria
```

```python
params = {'criterion' : ["gini", "entropy"],
         'splitter' : ["best", "random"],
          'max_depth' : [2, 4, 6],
          'max_features' : ["auto", "sqrt", "log2", None]
         }
rsv = RandomizedSearchCV(model, param_distributions=params, cv = 10, n_jobs=-1)
rsv.fit(X_train, y_train)
rsv.best_params_
```

```python
model.set_params(splitter='best',
                 max_features= 'auto',
                 max_depth= 6,
                 criterion= 'gini', random_state = 101)
model.fit(X_train, y_train)
```

```python
#Predicitons
y_pred = model.predict(X_test)
```

```python
#Model Accuracy
print('test score - ', str(model.score(X_test, y_test)))
print('train score - ', str(model.score(X_train, y_train)))
```

```python
#Confusion Matrix
ax = sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cbar = False)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus
import graphviz
```

```python
plt.figure(figsize=(12, 8))
dot_data = StringIO()
export_graphviz(model,
                out_file=dot_data,
                filled=True,
                rounded=True,
                special_characters=True,
                feature_names=['Age','EstimatedSalary','Gender_Male'],
                class_names=['0','1']
               )
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('ads2.png')
Image(graph.create_png())
```

```python
model2 = DecisionTreeClassifier(splitter='best',
                 max_features= 'auto',
                 max_depth= 6,
                 criterion= 'entropy', random_state = 101)#Using entropy as out splitting criteria
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

As we can see accuracy got decreased a bit when we use entropy as our splitting criteria

```python
dot_data=StringIO()
export_graphviz(model2,out_file=dot_data,rounded=True,filled=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("ads3.png")
Image(graph.create_png())
```
